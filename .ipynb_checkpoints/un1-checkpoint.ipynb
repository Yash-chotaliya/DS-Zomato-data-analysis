```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./dataset/zomato.csv',encoding='latin-1')
df.head()
```

```python
print(df.columns)
print('\n')
print(df.info())
df.describe()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(15,5)
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df_country=pd.read_excel('./dataset/Country-Code.xlsx')
df_country
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
country_name = final_df.Country.value_counts().index
country_count = final_df.Country.value_counts().values
```

```python
plt.pie(country_count[:3],labels=country_name[:3],autopct='%.1f%%')
```

observation : from above chart we can say that maximum number of transaction happened in india

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings.head()
```

```python
sns.barplot(x='Aggregate rating', y='Rating count', hue='Rating color', data = ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
# find the country that has 0 rating

final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename(columns={0:'Rating count'})
```

```python
# find out which currency is used by which country

final_df.groupby(['Country','Currency']).size().reset_index().drop(0,axis=1)
```

```python
final_df.columns
```

```python
# which country has online delivery option 

final_df[final_df['Has Online delivery']=='Yes'].groupby('Country').size()
```

observation : online delivery is available in India and UAE

```python
# draw a pie chart for city distribution

city_name = final_df.City.value_counts().index
city_count = final_df.City.value_counts().values

plt.pie(city_count[:5],labels=city_name[:5],autopct='%.2f%%')
```
